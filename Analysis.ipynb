```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Ahmedabad Properties.xlsx")
```

```python
df.head()
```

```python
df["location"].unique()
```

```python
df["location"].value_counts()
```

```python
df["location"]=df["location"].str.replace("ahmedabad","").str.replace(",","")
avg_price=df.groupby('location')['price_lakhs'].mean()
```

```python
avg_price
```

```python
plt.figure(figsize=(10, 20))  # Width x Height
plt.barh(avg_price.index, avg_price.values, height=0.5)
plt.xlabel("Properties price in lakhs")
plt.ylabel("Area")
plt.title("Average Price In Area",size=28,color="blue")
plt.show()
```

```python
num_properties=df["location"].value_counts()
```

```python
top_10=num_properties[:10]
```

```python
top_10
```

```python
plt.figure(figsize=(8,8))
plt.pie(top_10.values, autopct='%1.1f%%', startangle=140)
plt.title("Properties Distribution by Location",size=24,color="blue")
plt.axis('equal')
plt.legend(top_10.index)
plt.show()
```

```python
top_5_properties=df["name"].value_counts()[:5]
```

```python
top_5_properties
```

```python
plt.figure(figsize=(20,10))
plt.pie(top_5_properties,autopct="%1.1f%%")
plt.title("Top 5 Properties by name",size=24)
plt.legend(top_5_properties.index)
plt.show()
```

```python
top10_max_area=df.groupby("location")["area_sqft"].max().sort_values(ascending=False)[:10]
top10_min_area=df.groupby("location")["area_sqft"].min().sort_values()[:10]
top10_avg_area=df.groupby("location")["area_sqft"].mean().sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(12,8))
plt.bar(top10_max_area.index,top10_max_area.values,color="blue")
plt.xlabel("Area Name")
plt.xticks(rotation=45)
plt.title("Top 10 Location With Maximum Area ",size=18)
plt.ylabel("Area SQFT")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(top10_min_area.index,top10_min_area.values,color="orange")
plt.xlabel("Area Name")
plt.xticks(rotation=45)
plt.title("Top 10 Location With Minumum Area ",size=18)
plt.ylabel("Area SQFT")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(top10_avg_area.index,top10_avg_area.values,color="brown")
plt.xlabel("Area Name")
plt.xticks(rotation=45)
plt.title("Top 10 Location With Average Area ",size=18)
plt.ylabel("Area SQFT")
plt.show()
```

```python
top_10_area_avg_price=df.groupby('location')["price_lakhs"].mean().sort_values(ascending=False)[:10]
```

```python
top_10_area_avg_price
```

```python
plt.figure(figsize=(12,8))
plt.bar(top_10_area_avg_price.index,top_10_area_avg_price.values,color="red")
plt.xlabel("Area Name")
plt.xticks(rotation=45)
plt.title("Top 10 Area By Average Price ",size=18)
plt.ylabel("Area Average Price in lakhs")
plt.show()
```

```python
sorted_avg_price=top_10_area_avg_price.sort_values()
```

```python
plt.figure(figsize=(10,6))
plt.plot(sorted_avg_price)
plt.xlabel("Area name")
plt.ylabel("Average Price")
plt.xticks(rotation=90)
plt.title("Top 10 Area By Average Price ",size=18)
plt.grid(True)
plt.show()
```
